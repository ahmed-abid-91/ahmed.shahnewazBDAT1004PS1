Question 3 C#

```
double triangleArea(int a, int b, int c)
{ 
double s = (a+b+c)/2;
double calc = s*(s-a)*(s-b)*(s-c);
double area = Math.Sqrt(calc);
    return area;
}
Console.WriteLine(triangleArea(2, 2, 2));
```

Question 4 C#

```
Console.WriteLine("Input the number of elements to be stored in the array:");
int n = Convert.ToInt32(Console.ReadLine());

int[] arr = new int[n];

Console.WriteLine($"Input {n} elements in the array");
for (int i = 0; i < n; i++) 
{
    Console.WriteLine($"element - {i} : ");
    arr[i] = Convert.ToInt32(Console.ReadLine());
}
int evenindex = 0;
int oddindex = 0;
int[] evenArr = new int[n];
int[] oddArr = new int[n];

foreach (int i in arr)
{
    if(i%2 == 0)
    {
        evenArr[evenindex] = i;
        evenindex++;
    }
    else
    {
        oddArr[oddindex] = i;
        oddindex++;
    }
}

Array.Resize(ref evenArr, evenindex);
Array.Resize(ref oddArr, oddindex);
Console.WriteLine("The even elements are");
printArray(evenArr);
Console.WriteLine("\n");
Console.WriteLine("The odd elements are");
printArray(oddArr);

void printArray(int[] arr)
{
    foreach (int i in arr) 
    {
        Console.Write(i + " ");
    }

}
```

```
// (A)

bool inside(double x, double y, double x1, double y1, double x2, double y2)
{
    if ((x>x1 && x<x2) && (y>y1 && y < y2))
    {
        return true;
    }
    else
    {
        return false;
    }
}

Console.WriteLine(inside(1, 1, 0, 0, 2, 3));
Console.WriteLine(inside(-1, -1, 0, 0, 2, 3));

// (B)

if(inside(1,1,0.3,0.5,1.1,0.7) && inside(1, 1, 0.5, 0.2, 1.1, 2))
{
    Console.WriteLine("Yes");
}
else
{
    Console.WriteLine("No");
}
```
